```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-pastel')
```

```python
results = pd.read_csv('results/22_06_30_09_47.csv')
results
```

```python
sub_epoch_size = 1/max(results['epoch'].value_counts())
sub_epoch_size
```

```python
results['sub_epoch'] = results.index * sub_epoch_size
results
```

```python
plt.figure(figsize=(13,7))
plt.title('Loss')
plt.plot(results.sub_epoch, results.d_loss)
plt.plot(results.sub_epoch, results.g_loss)
plt.xlabel('Epoch')
plt.ylabel('BCE Loss')
plt.legend(['Discriminator', 'Generator'])
plt.show()
```

```python
# every 20 rows
R = results[results.index % 20 == 0]
plt.figure(figsize=(13,7))
plt.title('Loss')
plt.plot(R.sub_epoch, R.d_loss)
plt.plot(R.sub_epoch, R.g_loss)
plt.xlabel('Epoch')
plt.ylabel('BCE Loss')
plt.legend(['Discriminator', 'Generator'])
plt.show()
```
